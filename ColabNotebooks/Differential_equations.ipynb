# Solving equations by hand

Consider a simple equation $\dot y(x) = x^2$ with initial conditions $y(0) = -1$. Let us solve it.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import gca as gca
from matplotlib.pyplot import grid as grid

y0 = -1

x0 = 0
xf = 10

Count = 1000
array_x = np.linspace(x0, xf, Count)
array_y = np.zeros(Count)

dx = (xf - x0) / Count

y = y0
for i in range(Count):
    x = array_x[i]

    dydx = x**2

    y = y + dydx*dx

    array_y[i] = y

plt.plot(array_x, array_y)
ax = gca()
# ax.set_aspect('equal', 'box')
grid(color='k', linestyle='--', linewidth=0.2)
```

Consider equations: 

$$\dot y(x) = -x, \ \ \  y(0) = 10$$
$$\dot y(x) = \cos(x), \ \ \  y(0) = 0$$
$$\dot y(x) = x^3-x, \ \ \ y(0) = 2$$
$$\dot y(x) = 10, \ \ \ y(0) = -5$$

```python
def solve(fnc, y0):

    x0 = 0
    xf = 10

    Count = 1000
    array_x = np.linspace(x0, xf, Count)
    array_y = np.zeros(Count)

    dx = (xf - x0) / Count

    y = y0
    for i in range(Count):
        x = array_x[i]

        dydx = fnc(x)

        y = y + dydx*dx

        array_y[i] = y

    plt.plot(array_x, array_y)
    ax = gca()
    # ax.set_aspect('equal', 'box')
    grid(color='k', linestyle='--', linewidth=0.2)
```

```python
fnc1 = lambda x : -x
y0 = 10
solve(fnc1, y0)
```

```python
from math import cos

fnc2 = lambda x : cos(x)
y0 = 0
solve(fnc2, y0)
```

```python
fnc3 = lambda x : x**3 - x
y0 = 2
solve(fnc3, y0)
```

```python
fnc4 = lambda x : 10
y0 = -5
solve(fnc4, y0)
```

# Solving more equations by hand

Consider an equation $\dot y(x) = -10 y$ with initial conditions $y(0) = 1$. Let us solve it. Let us prove that the solution is equivalent to $y(x) = e^{-10x}$

```python
def solve(fnc, y0, array_x):

    Count = array_x.shape[0]

    array_y = np.zeros(Count)

    dx = (xf - x0) / Count

    y = y0
    for i in range(Count):
        x = array_x[i]

        dydx = fnc(y, x)

        y = y + dydx*dx

        array_y[i] = y

    plt.plot(array_x, array_y)
    ax = gca()
    # ax.set_aspect('equal', 'box')
    grid(color='k', linestyle='--', linewidth=0.2)
```

```python
from numpy import exp

x0 = 0
xf = 1
Count = 1000
array_x = np.linspace(x0, xf, Count)

fnc = lambda y, x : -10*y
y0 = 1
solve(fnc, y0, array_x)

analytic_fnc = lambda x : exp(-10*x)
plt.plot(array_x, analytic_fnc(array_x), linestyle=':', linewidth=4)
```

Consider an equation $\dot y(x) = -y + x$ with initial conditions $y(0) = 10$. Let us solve it.

## scipy odeint

```python
from scipy.integrate import odeint

x0 = 0
xf = 12
Count = 1000
array_x = np.linspace(x0, xf, Count)

fnc = lambda y, x : -y + x
y0 = 10

array_y = odeint(fnc, y0, array_x)

plt.plot(array_x, array_y)
ax = gca()
# ax.set_aspect('equal', 'box')
grid(color='k', linestyle='--', linewidth=0.2)
```
